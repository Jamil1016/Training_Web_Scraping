```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
table_titles = soup.find_all('tr')[0]
table_columns = [row.text.strip() for row in table_titles.find_all('th')]
table_columns.remove('Ref.')
```

```python
df = pd.DataFrame(columns= table_columns)
df
```

```python
table = soup.find_all('table')[0]
table
```

```python
column_data = table.find_all('tr')[2:]

for rows in column_data:
    row_data = [row.text.strip() for row in rows.find_all(['th','td'])]
    print(row_data)
```
